```python
import pandas as pd
```

```python
df = pd.read_csv("http://51.91.120.89/extras/nba.csv", index_col='Name')
df
```

```python
# looking for NaN values
df.isna().sum()
```

```python
# remove rows/columns based on NaN values
df.dropna(axis=0, how='all', inplace=True)
df
```

```python
df.isna().sum()
```

```python
# replace NaN with empty string in College column
df["College"].replace(pd.np.NaN, "", inplace=True)
```

```python
df.isna().sum()
```

```python
df
```

```python
# convert pounds -> kilograms
df.Weight /= 2.2 
```

```python
# convert feet - inches format to cm
def convert_height(s):
    feet, inches = s.split("-")
    return int(feet)*30.4 + int(inches)*2.54
```

```python
df.Height = df.Height.apply(convert_height)
```

```python
df.to_csv("nba.csv", sep="|", index_label="nba_player")
```

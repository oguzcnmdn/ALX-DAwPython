```python
import pandas as pd
```

```python
df = pd.read_csv("http://51.91.120.89/extras/world.csv")
df.drop(columns="Unnamed: 0", inplace=True)
df
```

```python
df.Year.unique()
```

```python
# show changes in "Happiness Score" over the years with pivot_table method
pd.pivot_table(df, index="Year", values="Happiness Score")
```

```python
df.query(" Year==2015 ")["Happiness Score"].mean()
```

```python
# show changes in "Happiness Score" over regions and years with pivot_table method
pd.pivot_table(df, index=["Region", "Year"], values="Happiness Score")
```

```python
pd.pivot_table(df, index="Region", values=["Family","Happiness Score"], columns="Year")
```

```python
df1 = pd.pivot_table(df, index="Region", values=["Family","Happiness Score"], columns="Year", aggfunc=[min,max])
df1
```

```python
df1.columns
```

```python
df1[ [('min','Family', 2016), ('max', 'Happiness Score', 2017) ] ]
```

```python
import numpy as np

df1 = pd.pivot_table(df, index="Region", values=["Family","Happiness Score"], columns="Year", 
                     aggfunc=[np.median, np.std])
df1
```

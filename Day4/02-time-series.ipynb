```python
df = pd.read_csv("https://stooq.com/q/d/l/?s=cdr&d1=20200101&d2=20220114&i=d")
df
```

```python
df.dtypes
```

```python
df.Date = pd.to_datetime(df.Date, format="%Y-%m-%d")
df.dtypes
```

```python
df.set_index("Date", inplace=True)
df.head()
```

```python
df.index
```

```python
df.Close.plot(color='red', grid=True, figsize=(16,7))
```

```python
df.loc["2021-12"].Close.plot(color='red', grid=True, figsize=(16,7))
```

```python
df.Close.plot(color='red', grid=True, figsize=(16,7))
```

```python
# calculate highest close quote in every year
df.Close.resample("YS").max().plot(kind='bar', color='red', grid=True, figsize=(16,7))
```

```python
# calculate lowest close quote in every quarter
df.Close.resample("Q").min().plot(kind='bar', color='red', grid=True, figsize=(16,7))
```

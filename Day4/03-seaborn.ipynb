```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("http://51.91.120.89/extras/gapminderDataFiveYear.csv").query(" year==2007 ")
df
```

```python
sns.lmplot(data=df, x="lifeExp", y="gdpPercap", fit_reg=False, aspect=2.5, hue="continent")
plt.grid()
plt.show()
```

```python
sns.lmplot(data=df, x="lifeExp", y="gdpPercap", fit_reg=False, aspect=0.7, col="continent")
plt.show()
```

```python
# classic line chart
df = pd.read_csv("https://covid19.who.int/WHO-COVID-19-global-data.csv")
df.sample(n=10)
```

```python
df.Date_reported = pd.to_datetime(df.Date_reported, format="%Y-%m-%d")
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(data=df.query("Country_code=='PL' and Date_reported>='2023-01-01'") , 
             x="Date_reported", y="New_cases", color="red" )

sns.lineplot(data=df.query("Country_code=='DE' and Date_reported>='2023-01-01'") , 
             x="Date_reported", y="New_cases", color="green" )

sns.lineplot(data=df.query("Country_code=='FR' and Date_reported>='2023-01-01'") , 
             x="Date_reported", y="New_cases", color="blue" )

plt.grid()
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(data=df.query("Country_code in ['PL','DE','FR'] and Date_reported>='2023-01-01'") , 
             x="Date_reported", y="New_cases", hue="Country_code" )


plt.grid()
```

```python
df = sns.load_dataset("car_crashes").sort_values("total", ascending=False)
df
```

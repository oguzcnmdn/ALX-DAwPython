```python
import pandas as pd
```

```python
data = [
    {"fname":"Jan", "lname":"Kowalski", "score": 4.76, "class": 4},
    {"fname":"Andrzej", "lname":"Nowak", "score": 4.0, "class": 5},
    {"fname":"Adam", "lname":"Zieliński", "score": 5.0, "class": 6},
    {"fname":"Jan", "lname":"Kowalski", "score": 4.76, "class": 4},
    {"fname":"Andrzej", "lname":"Nowak", "score": 4.0, "class": 5},
    {"fname":"Jan", "lname":"Pawlak", "score": 5, "class": 6},
    {"fname":"Jan", "lname":"Kowalski", "score": 4.76, "class": 4},
]

df = pd.DataFrame(data)
df
```

```python
# how many duplicates?
df.duplicated().sum()
```

```python
# show duplicated rows
df[ df.duplicated() ]
```

```python
# drop duplicated rows
df.drop_duplicates(keep="first")
```

```python
# drop duplicated rows
df.drop_duplicates(keep="last")
```

```python
# drop duplicated rows
df.drop_duplicates(keep=False)
```

```python
# drop duplicates based on column/columns
df.drop_duplicates(subset=["fname"])
```

```python
df1 = pd.read_csv("http://51.91.120.89/extras/otodom-head-sample.csv", sep=";").iloc[ : , 3:]
df1.head()
```

```python
df2 = pd.read_csv("http://51.91.120.89/extras/otodom-details-sample.csv", sep=";", usecols=["ID","descr"])
df2.head()
```

```python
df1.shape, df2.shape
```

```python
pd.merge(df1, df2, how="inner", on="ID") # left_on="ID", right_on="ID"
```

```python
# left join
df3 = pd.merge(df1, df2, how="left", on="ID", indicator=True)
df3
```

```python
df3._merge.value_counts()
```

```python
df3.query(" _merge=='left_only'  ")
```

```python
# right join
df3 = pd.merge(df1, df2, how="right", on="ID", indicator=True)
df3
```

```python
df3._merge.value_counts()
```

```python
df1 = df1.sample(frac=0.25, random_state=0)
df2 = df2.sample(frac=0.25, random_state=0)
```

```python
df1.shape, df2.shape
```

```python
# outer join
df3 = pd.merge(df1, df2, how="outer", on="ID", indicator=True)
df3
```

```python
df3._merge.value_counts()
```

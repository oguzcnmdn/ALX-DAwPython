```python
import pandas as pd
```

```python
df = pd.read_csv("http://51.91.120.89/extras/drinksbycountry.csv")
df
```

```python
df.rename(columns={"total_litres_of_pure_alcohol":"total"}, inplace=True)
df.head(10)
```

```python
# grouping data base on one column
dfg = df.groupby("continent")
```

```python
# get numbers of items(rows) in every bucket
dfg.size()
```

```python
dfg.count()
```

```python
# average consumption for every continent
dfg.mean()
```

```python
dfg.max()
```

```python
dfg[ ["beer_servings","total"] ].max()
```

```python
dfg.max()[ ["beer_servings","total"] ]
```

```python
# apply max() and min() for every continent
df_mm = dfg[ ["beer_servings","total"] ].agg([min, max])
```

```python
df_mm.columns
```

```python
df_mm[ [('beer_servings', 'min'), ('total', 'max')] ]
```

```python
# looking for one country with highest consumption (based on total column) from every continent
df.sort_values("total", ascending=False).groupby('continent').first()
```

```python
# looking for TOP3 countries with highest consumption (based on total column) from every continent
df.sort_values("total", ascending=False).groupby('continent').head(3).sort_values(["continent","total"], ascending=[True, False])
```
